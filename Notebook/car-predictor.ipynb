### Cleaning Data

```python
## Cleaning Price

car['Price'].unique()
```

```python
print(car.shape)
print(car[car['Price'] == 'Ask For Price']['Price'].count())
car = car[car['Price'] != 'Ask For Price']
print(car.shape)
car['Price'] = car['Price'].str.replace(',', '').astype(int)
print(car.info())
```

```python
car['Price'].unique()
```

```python
## Year Cleaning

car['year'].unique()
```

```python
car= car[car['year'].str.isnumeric()]
```

```python
car['year'] = car['year'].astype(int)
```

```python
car.shape
```

```python
## kms driven
car['kms_driven'].unique()
```

```python
car['kms_driven'] = car['kms_driven'].str.split().str.get(0).str.replace(',', '')
```

```python
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car.shape
```

```python
car.info()
```

```python
car
```

```python
car['fuel_type'].unique()
```

```python
car = car[~car['fuel_type'].isna()]
```

```python
car.shape
```

```python
car['company'].unique()
```

```python
car['name'].unique()
```

```python
car['name'] = car['name'].str.split().str.slice(0, 3).str.join(' ')
```

```python
car.describe()
```

```python
car = car[car['Price'] < 6000000]
```

```python
car = car.reset_index(drop=True)
```

### Model Training


```python
X = car.drop(columns='Price', axis=1)
y = car['Price']
X.shape , y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

cat_features = X.select_dtypes(include='O').columns
num_features = X.select_dtypes(exclude='O').columns

ohe = OneHotEncoder()
ohe.fit(X[cat_features])

transformer = ColumnTransformer([
    ('column_transformer', OneHotEncoder(categories=ohe.categories_) , cat_features),
    ('standard_scaler', StandardScaler(), num_features)
], remainder='passthrough')

lr = LinearRegression()

pipe = Pipeline([('preprocessor', transformer), ('regressor', lr)])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_train)
from sklearn.metrics import r2_score
print(r2_score(y_train, y_pred))
```

```python
y_test_pred = pipe.predict(X_test)
```

```python
print(r2_score(y_test, y_test_pred))
```

```python
from sklearn.model_selection import train_test_split

scores = []
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
    pipe = Pipeline([('preprocessor', transformer), ('regressor', LinearRegression())])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    scores.append(r2_score(y_test, y_pred))
```

```python
np.argmax(scores)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=124)
pipe = Pipeline([('preprocessor', transformer), ('regressor', LinearRegression())])
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test, y_pred))
```
